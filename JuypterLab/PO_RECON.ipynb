```python
cspoms_oracle_query = '''
SELECT 
    xph.po_num,
    xph.po_selection_fac AS facility,
    xph.po_date_ordered AS order_date,
    xph.vendor_number AS vendor_num,
    xph.po_need_by_date AS due_date,
    xph.po_rec_type AS perish_po,
    xph.po_type
    ,cb.biceps_buyer_id AS bic_buyer_id
    ,xph.received_date as po_received
    ,xph.po_transfer_flag as pullback_flag
    --as po_oss_dest
    ,xph.secondary_po_num as secondary_po
    ,'CSPOMS' as recon_origin
    ,'CSPOMS' as inbound_source

FROM 
    xxpo_po_headers xph
inner JOIN 
    cspoms_replenishment_config cfg ON xph.po_selection_fac = cfg.facility_code
    AND cfg.facility_type = 'Oracle' 
LEFT JOIN 
    XXPO_CUSTOM_BUYERS cb ON xph.po_buyer_number = cb.buyer_number
    AND cb.end_date IS NULL  
WHERE 
    xph.po_status = 'Open' 
    AND xph.received_date IS NULL;
'''

post_cur.execute(cspoms_oracle_query)
c_column_names = [desc[0] for desc in post_cur.description]
cspoms_oracle_df = post_cur.fetchall()
cspoms_oracle_df = pd.DataFrame(cspoms_oracle_df, columns=c_column_names)

cspoms_oracle_df
```

```python
from psycopg2.extras import execute_batch
post_conn = connection

def insert_data_in_batches( dataframe, connection, cursor, table_name):
    num_records = len(dataframe)
    fields = [col for col in dataframe.columns]
    str_sub = ', '.join(['%s'] * len(fields))
    insert_query = f"INSERT INTO {table_name} ({', '.join(fields)}) VALUES ({str_sub})"

    try:
        print('before insert')
        datadict = dataframe.to_dict('records')
        tuple_rec = [tuple(df.values()) for df in datadict]
        execute_batch(cursor, insert_query, tuple_rec)
        print('after insert')
    except Exception as e:
        connection.rollback()
        print(f'Error inserting batch: {e}')
        raise

    print(f'INSERTED TOTAL {num_records} RECORDS TO {table_name} TABLE')
    
try:
    print(f'Length of DataFrame:', len(po_not_in_oracle))
    insert_data_in_batches(po_not_in_oracle, post_conn, post_cur, 'XXPO_PO_RECON_INBOUND')
    print('Data insertion completed successfully.')
except Exception as e:
    print('Error during data insertion:', e)
print('data_tuples prepared')
post_conn.commit()
```
